```python
import pandas as pd

zip_df = pd.read_csv('Zip_zori_sm_month.csv')

zip_bool = zip_df['City'] == 'Boston'

zip_df = zip_df.loc[zip_bool, :]

zip_df
```

```python
zip_df.reset_index(drop = True, inplace = True)

zip_df

#zip_df.to_csv('zip_df.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(zip_df.iloc[1, 9:])
```

```python
from sklearn import linear_model
import numpy as np

rent_dict = {}
for idx, row in zip_df.iterrows():
    x = [[i] for i in range(0, 94) if row[i+9] == row[i+9]]
    y = [[row[i]] for i in range(9, 103) if row[i] == row[i]]
    
    linear = linear_model.LinearRegression()
    linear.fit(x, y)

    rent_dict['0'+str(row['RegionName'])] = [row[102]+(linear.coef_[0][0]*i) for i in range(1, 13)]

rent_dict
```

```python
zips = input('What zip codes are you looking to live in? ')
zips = zips.replace(' ', '').split(',')

month_dict = {'january': 0, 'february': 1, 'march':2, 'april': 3, 'may':4, 'june':5, 'july':6, 'august':7, 'september':8, 'october':9, 'november':10, 'december':11}

beg_month = input('What is the first month in 2023 you want to lease during? ').lower()
assert beg_month in month_dict.keys(), 'invalid month input'

end_month = input('What is the last month in 2023 you want to lease during? ').lower()
assert end_month in month_dict.keys(), 'invalid month input'

start = month_dict[beg_month]
end = month_dict[end_month]

new_dict = {}
for key, value in rent_dict.items():
    if key in zips:
        new_dict[key] = [value[start], value[end], sum(value[start:end])/len(value[start:end])]
        
beg_min = ''
end_min = ''
mean_min = ''
for key, value in new_dict.items():
    if value[0] == min(new_dict.values())[0]:
        beg_min = key
    if value[1] == min(new_dict.values())[1]:
        end_min = key
    if value[2] == min(new_dict.values())[2]:
        mean_min = key

print(f'\nThe zip code with the lowest price at the beginning of your 2023 leasing period is {beg_min} (${new_dict[beg_min][0].round(2)}).\nThe zip code with the lowest price at the end of your 2023 leasing period is {end_min} (${new_dict[end_min][1].round(2)}).\nThe zip code with the lowest price average price during your 2023 leasing period is {mean_min} (${new_dict[mean_min][2].round(2)}).')

graph = input('\nWould you like to see how rent prices changed in Boston during 2022?(yes/no) ')
if graph == 'yes' or graph == 'Yes':
    sns.set()

    for idx, row in zip_df.iterrows():
        if '0'+str(row['RegionName']) in zips:
            lw = 5
        else:
            lw = 1
        plt.plot(row[91:103], label = '0'+str(row['RegionName']), linewidth = lw)
        

    plt.legend(bbox_to_anchor=(1.1, 1))
    plt.gcf().set_size_inches(15, 9)
    plt.xlabel('Date')
    plt.ylabel('Average Rent Price')
    plt.title('Average Rent Prices in Boston (2022)')
    plt.show()
else:
    pass
```
